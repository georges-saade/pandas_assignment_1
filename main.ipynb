## Task 4

Read about data types, specifically in Pandas. Write examples of how to change from one data type to another using the dataframes created above

The different data types in pandas are:
- Object: refers usually to text or mixed numeric and non-numeric values
- int64: integer numbers
- float64: float numbers
- bool: bool values (True or False)
- datetime64: Date and Time values
- timedelta[ns]: Difference between 2 datetimes
- category: Finite list of text values

To convert a pandas column of data in a dataFrame to a different type we use astype() function.

In the exmaple 2 above we will try to convert the rating column into an integer

```python
df2["Rating"].astype('int')
```

Since the astype function returns a copy we assign it back to the df2 dataFrame

```python
df2["Rating"]=df2["Rating"].astype('int')
```
